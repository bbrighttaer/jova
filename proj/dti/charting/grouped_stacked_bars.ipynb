```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
filename = 'kiba_jova'
df = pd.read_csv(f'{filename}.csv')
```

```python
df.head()
```

```python
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('split:N', title=None, axis=alt.Axis(offset=8.0, orient='bottom')),
    y=alt.Y('value:Q', axis=alt.Axis(grid=False, title=None,offset=8.0, orient='left')),
    column=alt.Column('model:N', title=None, spacing=65),
    color=alt.Color('metric:N',
            scale=alt.Scale(range=['#96ceb4', '#ffcc5c','#ff6f69'],),
        )).configure().configure_view(
        strokeOpacity=0,
        strokeCap = 'round'
    ).configure_axis(tickBand='extent')
chart.save(f'{filename}.png', scale_factor=3.0)
chart.display()
```
